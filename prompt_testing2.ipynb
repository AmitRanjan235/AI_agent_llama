```python
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import UnstructuredWordDocumentLoader
import os, getpass
from langchain.schema import Document
import pandas as pd

# Setup Groq LLM
os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API Key: ")
llm = ChatGroq(model_name="llama3-70b-8192", groq_api_key=os.environ["GROQ_API_KEY"])
```

```python
import os
from google.colab import drive

os.makedirs("data", exist_ok=True)
drive.mount('/content/drive')
```

```python
# ---- 1. Load DOCX files ----
docx_paths = [
    "/content/drive/MyDrive/rbi_docs/doc1.docx",
    "/content/drive/MyDrive/rbi_docs/doc2.docx",
    "/content/drive/MyDrive/rbi_docs/doc3.docx"
]

documents = []
for path in docx_paths:
    loader = UnstructuredWordDocumentLoader(path)
    documents.extend(loader.load())

# ---- 2. Load Excel (flow/subflow) ----
excel_path = "/content/drive/MyDrive/rbi_docs/Demo.xlsx"   # <--- change to your file
df = pd.read_excel(excel_path)

# **Assume columns: "State", "AgentReply", "NextStep" (update if different)**
flow_docs = []
for i, row in df.iterrows():
    doc_text = f"State: {row.get('State','')} | Agent Reply: {row.get('AgentReply','')} | Next Step: {row.get('NextStep','')}"
    flow_docs.append(Document(page_content=doc_text, metadata={"source": "flow_excel", "row": i}))

# ---- 3. Chunk text documents (DOCX) ----
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
text_chunks = text_splitter.split_documents(documents)

# ---- 4. Combine DOCX text chunks and Excel subflow docs ----
all_documents = text_chunks + flow_docs

# ---- 5. Build or load embeddings as usual ----
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

faiss_path = "/content/drive/MyDrive/rbi_docs/faiss_index_rbi2"
if os.path.exists(faiss_path):
    docsearch = FAISS.load_local(faiss_path, embeddings, allow_dangerous_deserialization=True)
    print("✅ Loaded FAISS index from disk")
else:
    docsearch = FAISS.from_documents(all_documents, embedding=embeddings)
    docsearch.save_local(faiss_path)
    print("✅ FAISS vector store created and saved successfully!")
```

```python
def pre_due_reminder_prompt(context, customer_name, amount, due_date, language, lender="Your Lender"):
    return f"""
You are a collections agent for {lender}. Your job is to send a gentle reminder to the customer about their upcoming EMI payment.

Customer details:
- Name: {customer_name}
- EMI amount: ₹{amount}
- Due date: {due_date}
- Language: {language}

Guidelines:
- Use a warm, respectful, and helpful tone.
- Address customer as "Sir/Ma'am" or by their preferred name with honorific ("ji").
- If in Hindi, use appropriate code-mixing and formal pronouns ("aap").
- Offer help if the customer needs assistance.
- Keep your response short and concise (2-3 sentences max).
- Example: "This is a friendly reminder that your loan installment of ₹{amount} is due on {due_date}. Please ensure timely payment to avoid late fees. Let us know if you need any help."

Context: {context}
Generate the reminder in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def post_due_prompt(context, customer_name, loan_type, overdue_amount, dpd, risk_category, language):
    return f"""
You are a collections agent following up with a customer whose payment is overdue.

Customer details:
- Name: {customer_name}
- Loan type: {loan_type}
- Amount overdue: ₹{overdue_amount}
- Days past due: {dpd}
- Risk category: {risk_category}
- Language: {language}

Guidelines:
- Use an assertive but polite tone.
- Mention overdue status and politely request payment.
- For early bucket (1–30 DPD): Focus on understanding the reason and being supportive.
- For mid/late buckets (31–90+ DPD): Increase urgency and mention possible consequences (late fees, impact on credit).
- Always stay compliant with RBI regulations.
- Offer solutions (partial payment, date negotiation) if customer cites hardship.
- Keep your response short and concise (2-3 sentences max).

Context: {context}
Generate the follow-up message in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def hardship_prompt(context, customer_name, reason, language):
    return f"""
The customer says they can't pay due to a specific reason: {reason}

Guidelines:
- Empathize sincerely ("I understand your situation, Sir/Ma'am.").
- Offer to find a solution: smaller payment, extension, or new plan.
- Use the customer's language and cultural context (urban/rural).
- Keep your response short and concise (2-3 sentences max).
- Example: "We understand that there is a curfew in your area. If needed, we can reschedule your payment or accept a partial amount. Let us know what would help you."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def partial_payment_prompt(context, customer_name, partial_amount, rest_date, language):
    return f"""
The customer cannot pay the full EMI but is willing to pay part.

Guidelines:
- Be collaborative and supportive.
- Suggest splitting the payment if possible.
- Confirm dates for remaining amount.
- Keep your response short and concise (2-3 sentences max).
- Example: "Would it be possible to pay ₹{partial_amount} this week, and the rest by {rest_date}? We want to help you maintain a good record."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def payment_confirmation_prompt(context, customer_name, pay_date, language):
    return f"""
The customer promises to pay on a specific date.

Guidelines:
- Thank them, confirm commitment, and politely remind of importance.
- Keep your response short and concise (2-3 sentences max).
- Example: "Thank you for your commitment to pay by {pay_date}. Please try to honor it. Let us know if any issue comes up."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def legal_escalation_prompt(context, customer_name, language):
    return f"""
The account is in severe default (90+ days). You must warn the customer of possible legal steps without threatening.

Guidelines:
- Be formal, clear, and respectful.
- State policy, not personal threat.
- Keep your response short and concise (2-3 sentences max).
- Example: "As per our policy and your loan agreement, legal recovery proceedings may be initiated if payment is not received within 7 days. We wish to avoid this and would appreciate your cooperation."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def callback_prompt(context, customer_name, language):
    return f"""
Customer says they are busy/unavailable.

Guidelines:
- Apologize for inconvenience.
- Offer to call back at their convenience.
- Confirm preferred time/date.
- Keep your response short and concise (2-3 sentences max).
- Example: "Sorry for the inconvenience, Sir/Ma'am. When would be a good time to call you back?"

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def info_query_prompt(context, customer_name, info_requested, language):
    return f"""
The customer asks for info on {info_requested}.

Guidelines:
- Give the requested info clearly.
- Offer to answer follow-up questions.
- Keep your response short and concise (2-3 sentences max).
- Example: "Your current outstanding principal is ₹X. The loan will end on Y. Current interest rate is Z%."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def complaint_prompt(context, customer_name, complaint, language):
    return f"""
Customer raises a complaint or dispute: {complaint}

Guidelines:
- Remain calm, patient, and solution-oriented.
- Assure the customer you will escalate or resolve.
- Keep your response short and concise (2-3 sentences max).
- Example: "Thank you for bringing this to our notice. I will escalate this to the concerned department and ensure you are updated."

Context: {context}
Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""

def glossary_prompt(term, language):
    return f"""
Customer asks "What does {term} mean?" or needs a definition.

Guidelines:
- Explain the term in simple, layperson-friendly language.
- Keep your response short and concise (2-3 sentences max).
- Example: "EMI means Equated Monthly Installment. It is the fixed amount you pay every month towards your loan."

Respond in {language}.
Only reply with the exact next message the agent should say (1-2 sentences). Do NOT include any explanations or apologies unless required by the situation.
"""
```

```python
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=docsearch.as_retriever(search_kwargs={'k': 3}),
    return_source_documents=True
)
```

```python
while True:
    print("\n--- New Query ---")
    print("Scenarios:")
    print("  pre_due    : Pre-due polite reminder")
    print("  post_due   : Overdue/late payment follow-up")
    print("  hardship   : Customer hardship (illness/job loss/curfew etc.)")
    print("  partial    : Partial payment negotiation")
    print("  confirm    : Payment commitment/confirmation")
    print("  legal      : Legal escalation (90+ DPD, warning)")
    print("  callback   : Callback/customer unavailable")
    print("  info       : Info query (principal, interest, etc.)")
    print("  complaint  : Complaint/dispute handling")
    print("  glossary   : Explain a term/jargon\n")
    scenario = input("Enter scenario: ").strip().lower()
    if scenario not in [
        "pre_due", "post_due", "hardship", "partial", "confirm", "legal", "callback", "info", "complaint", "glossary"
    ]:
        print("❌ Invalid scenario. Try again.")
        continue

    user_query = input("Enter user query or question: ").strip()
    language = input("Enter language (English/Hindi/Marathi/Tamil): ").strip()

    # Retrieve context from FAISS index (using the user's question)
    context_docs = qa.retriever.get_relevant_documents(user_query)
    context = " ".join([doc.page_content for doc in context_docs])

    # Fill out prompt based on scenario
    if scenario == "pre_due":
        customer_name = input("Customer Name: ").strip()
        amount = input("EMI Amount: ").strip()
        due_date = input("Due Date: ").strip()
        prompt = pre_due_reminder_prompt(context, customer_name, amount, due_date, language)
    elif scenario == "post_due":
        customer_name = input("Customer Name: ").strip()
        loan_type = input("Loan Type: ").strip()
        overdue_amount = input("Overdue Amount: ").strip()
        dpd = input("Days Past Due: ").strip()
        risk_category = input("Risk Category: ").strip()
        prompt = post_due_prompt(context, customer_name, loan_type, overdue_amount, dpd, risk_category, language)
    elif scenario == "hardship":
        customer_name = input("Customer Name: ").strip()
        reason = input("Reason for Delay: ").strip()
        prompt = hardship_prompt(context, customer_name, reason, language)
    elif scenario == "partial":
        customer_name = input("Customer Name: ").strip()
        partial_amount = input("Partial Amount: ").strip()
        rest_date = input("Remaining Date: ").strip()
        prompt = partial_payment_prompt(context, customer_name, partial_amount, rest_date, language)
    elif scenario == "confirm":
        customer_name = input("Customer Name: ").strip()
        pay_date = input("Commitment Date: ").strip()
        prompt = payment_confirmation_prompt(context, customer_name, pay_date, language)
    elif scenario == "legal":
        customer_name = input("Customer Name: ").strip()
        prompt = legal_escalation_prompt(context, customer_name, language)
    elif scenario == "callback":
        customer_name = input("Customer Name: ").strip()
        prompt = callback_prompt(context, customer_name, language)
    elif scenario == "info":
        customer_name = input("Customer Name: ").strip()
        info_requested = input("Info Requested (principal/interest/etc): ").strip()
        prompt = info_query_prompt(context, customer_name, info_requested, language)
    elif scenario == "complaint":
        customer_name = input("Customer Name: ").strip()
        complaint = input("Complaint: ").strip()
        prompt = complaint_prompt(context, customer_name, complaint, language)
    elif scenario == "glossary":
        term = input("Term to define: ").strip()
        prompt = glossary_prompt(term, language)

    print("\n🟩 Querying LLM...")
    response = llm.invoke(prompt)
    # Show only the direct agent reply (1-2 lines, as you want)
    print("\n==== Agent Response ====\n")
    print(response.content.strip())
    print("\n========================\n")
```
